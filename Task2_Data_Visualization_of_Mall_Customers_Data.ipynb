## Data Visualization of Mall Customers Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'E:\ExcelR\Data Science\Classroom Activities\November_23\Datasets\Mall_Customers.csv')
df.describe(), df.info()
```

```python
df.head()
```

### Line Chart

```python
df['AnnualIncome(k$)'].plot()
```

```python
df['SpendingScore(1-100)'].plot()
```

```python
df[['AnnualIncome(k$)', 'SpendingScore(1-100)']].plot.line()
```

### Scatter Chart

```python
df.plot(kind = 'scatter', x = 'Age', y = 'AnnualIncome(k$)')
df['Age'].corr(df['AnnualIncome(k$)'])
```

```python
df.plot(kind = 'scatter', x = 'Age', y = 'SpendingScore(1-100)')
df['Age'].corr(df['SpendingScore(1-100)'])
```

### Histogram

```python
df['AnnualIncome(k$)'].plot(kind = 'hist', bins = 20)
```

```python
df['SpendingScore(1-100)'].plot(kind = 'hist', bins = 20)
```

```python
df[['AnnualIncome(k$)', 'SpendingScore(1-100)']].plot(kind = 'hist', bins = 10, alpha = 0.5)
```

```python
gen_income = round(df.groupby("Gender")['AnnualIncome(k$)'].mean(), 2)
print(gen_income)

# bar chart
gen_income.plot.bar()
```

```python
gen_spend = round(df.groupby("Gender")['SpendingScore(1-100)'].mean(), 2)
print(gen_spend)

# bar chart
gen_spend.plot.bar()
```

```python
gen_grp_hist = round(df.groupby("Gender")[['AnnualIncome(k$)', 'SpendingScore(1-100)']].mean(), 2)
print(gen_grp_hist)

gen_grp_hist.plot(kind = 'hist', alpha = 0.6)
```

```python
df[['Age', 'AnnualIncome(k$)', 'SpendingScore(1-100)']].plot(kind = 'hist', bins = 30, alpha = 0.5)
```
